### Exercise 02.1
- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

<span style="color:red">Show a picture of both your estimations of $I$</span> and their uncertainties with a large number of *throws* $M$ (e.g. $M\ge 10^4$) as a function of the number of blocks, $N$</font>.

As an approximation we use the straight line between $(0,y(0))$ and $(1,y(1))$: $f(x)=\pi/2-x\pi/2$, since using a graphic calculator it looks like it follow the integrand more closely than the first order Taylor expansion in both 0 (which is y=1, so equal to part 1) and 1. To get a probability distribution we normalize it in the interval $[0,1]$, getting $p(x) = 2(1-x)$. Lastly to generate numbers following the probability distribution $p(x)$ we use its quantile, which is the inverse of the integral function in the interval $[0,1]$: $Q=1-\sqrt{1-x}$.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

# Read the file into a pandas DataFrame
Int1R = pd.read_csv('Int1.dat', header=None, sep='   ', engine='python')
Int2R = pd.read_csv('Int2.dat', header=None, sep='   ', engine='python')

# Extract the values and errors as separate Series
Int1 = Int1R[0].tolist()
Int1Err = Int1R[1].tolist()
Int2 = Int2R[0].tolist()
Int2Err = Int2R[1].tolist()

M=100000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N

x = np.arange( 1, len(Int1)+1)       # [1,2,...,N]
x*=L # Number of throws = block * (Number of throws in each block)

plt.errorbar(x, [val - 1 for val in Int1], yerr=Int1Err, label='Uniform distribution')
plt.errorbar(x, [val - 1 for val in Int2], yerr=Int2Err, label='Importance sampling with p(x)=2(1-x)')
plt.xlabel('#throws')
plt.ylabel('<I>-1')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Read the file into a pandas DataFrame
discR = pd.read_csv('disc.dat', header=None, sep='   ', engine='python')

# Extract the values and errors as separate Series
disc = discR[0].tolist()
discErr = discR[1].tolist()

x = np.arange( 1, len(disc)+1)       # [1,2,...,N]

plt.errorbar(x, disc, yerr=discErr)
plt.xlabel('#steps')
plt.ylabel('sqrt(<|r|^2>)')
plt.title('Cubic lattice with $a = 1$')
plt.grid(True)
plt.show()
```
