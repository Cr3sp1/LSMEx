Use the MD/MC code to <span style="color:red">print out instantaneous values of potential energy per particle, $U/N$, along **equilibrated** Monte Carlo simulations</span> (of $M=5\times 10^5$ MC steps and where the Metropolis acceptance ratio has been fixed approximately to 50%) in the following thermodynamic conditions:

1. <span style="color:red">Once this data has been produced, compute and make pictures with a python script of the autocorrelation functions of $U/N$ to approximately estimate how large the blocks in your simulations should be.

### <span style="color:red"> Note</span>
To calculate the autocorrelation efficiently we implemented an algorithm that uses the FFT in the following manner: </span>

1)Subtract the mean of the time series from each data point to center it around zero.

2)Compute the Fourier transform of the centered time series using np.fft.fft().

3)Compute the power spectrum by taking the absolute value of the Fourier transform and squaring it.

4)Compute the inverse Fourier transform of the power spectrum using np.fft.ifft().

5)Normalize the result by dividing by the length of the time series.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

UnSol = np.loadtxt('output_solid_UN.data')
UnGas = np.loadtxt('output_gas_UN.data')
UnLiq = np.loadtxt('output_liquid_UN.data')

UnSolVal = UnSol[:, 1]
UnLiqVal = UnLiq[:, 1]
UnGasVal = UnGas[:, 1]

def autocorr_fft(x):
    # Compute autocorrelation using FFT
    x = x - np.mean(x)                             #Subtracts the average so that the average is zero
    f = np.fft.fft(x)                              #Calculates the DFT of the data
    acf = np.real(np.fft.ifft(f * np.conj(f)))     #Takes the real part of the inverse DFT of the power spectrum of the data (the magnitude squared of f) whi
    acf /= len(x)                                  #Normalizes it
    return acf[:len(x)//2]

def plot_autocorr(data, name, colour, maxlag, maxauto):
    # Compute the autocorrelation function using FFT
    autocorr = autocorr_fft(data)

    # Plot the autocorrelation function
    plt.plot(autocorr, color=f'{colour}')
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.title(f'{name}')
    plt.grid(True)
    ax = plt.gca()
    ax.set_xlim([0, maxlag])
    ax.set_ylim([-0.0002, maxauto])
    plt.show()
    
    
plot_autocorr(UnSolVal, 'Solid', 'C0', 300, 0.001)
plot_autocorr(UnLiqVal, 'Liquid', 'C1', 600, 0.001)
plot_autocorr(UnGasVal, 'Gas', 'C2', 100, 0.0004)
```

2. Use the same data to study, with a python script, the estimation of the statistical uncertainties of the expectation value of $U/N$ for different size of the blocks in the data blocking technique, from $L=M/N=10$ to $L=5\times 10^3$, being $N$ the number of blocks. <span style="color:red">Make pictures of these statistical uncertainties as a function of $L$.</span> What you observe in the statistical uncertainties after that the block size is large enough (from the point of view of the data blocking technique)?

```python
def data_blocking(data, block_sizes):
    """
    Computes the statistical uncertainties of the expectation value of the data for different block sizes.
    
    Parameters:
    data (array): Input data.
    block_sizes (array): Array of block sizes.
    
    Returns:
    Tuple of two arrays: Block sizes and corresponding statistical uncertainties.
    """
    n = len(data)
    num_blocks = len(block_sizes)
    uncertainties = np.zeros(num_blocks)
    
    for i in range(num_blocks):
        L = block_sizes[i]
        num_blocks_L = n // L
        block_means = np.zeros(num_blocks_L)
        
        for j in range(num_blocks_L):
            block_means[j] = np.mean(data[j*L:(j+1)*L])
            
        mean_L = np.mean(block_means)
        mean_sq_L = np.mean(block_means**2)
        sigma_L = np.sqrt((mean_sq_L - mean_L**2) / (num_blocks_L - 1))
        uncertainties[i] = sigma_L
    
    return block_sizes, uncertainties


block_sizes = np.arange(10, 5001, 10)
bs, uncs = data_blocking(UnSolVal, block_sizes)

plt.plot(bs, uncs)
plt.xlabel('Block size')
plt.ylabel('Statistical uncertainty on U/N')
plt.title('Solid')
plt.grid(True)
plt.show()

bs, uncs = data_blocking(UnLiqVal, block_sizes)

plt.plot(bs, uncs, color = 'C1')
plt.xlabel('Block size')
plt.ylabel('Statistical uncertainty on U/N')
plt.title('Liquid')
plt.grid(True)
plt.show()


bs, uncs = data_blocking(UnGasVal, block_sizes)

plt.plot(bs, uncs, color = 'C2')
plt.xlabel('Block size')
plt.ylabel('Statistical uncertainty on U/N')
plt.title('Gas')
plt.grid(True)
plt.show()
```

### Results
Our resluts suggest that the statistical uncertainty stops growing at around a block size of 1000-2000 for the liquid and before that point in the other two cases. To be safe we decided to use a block size of 3000.

show in pictures the obtained average values and uncertainties for the potential energy per particle, $U/N$, the pressure $P$ and the radial distribution function $g(r)$ in LJ units ... <span style="color:red"> and compare your MC results for the radial distribution function, $g(r)$, with those obtained with Molecular Dynamics NVE simulations in similar thermodynamic conditions.</span>

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def read_data(filename):
    # Read the data from file
    data = np.loadtxt(filename)
    # Extract the values and errors as separate Series
    x = data[:, 0]
    val = data[:, 2]
    err = data[:, 3]
    return x, val, err

def plot_data(x, y1_val, y1_err, y1_label, y2_val, y2_err, y2_label, name, ylable ):
    # Set up the plot
    plt.errorbar(x, y1_val, yerr=y1_err, label=y1_label)
    plt.errorbar(x, y2_val, yerr=y2_err, label=y2_label, alpha=0.7)
    plt.xlabel('#throws')
    plt.ylabel(f'{ylable}')
    plt.title(f'{name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Read the data for the solid phase
x, EnSolMCVal, EnSolMCErr = read_data('output_solid_epot_MC.dat')
_, EnSolMDVal, EnSolMDErr = read_data('output_solid_epot_MD.dat')

# Calculate the number of throws
L = 1000
N = 50
M = L * N
x *= L

# Plot the solid phase data
plot_data(x, EnSolMCVal, EnSolMCErr, "Solid MC", EnSolMDVal, EnSolMDErr, "Solid MD", "U/N", "U*")

# Read the data for the liquid phase
x, EnLiqMCVal, EnLiqMCErr = read_data('output_liquid_epot_MC.dat')
_, EnLiqMDVal, EnLiqMDErr = read_data('output_liquid_epot_MD.dat')

# Calculate the number of throws
M = L * N
x *= L

# Plot the liquid phase data
plot_data(x, EnLiqMCVal, EnLiqMCErr, "Liquid MC", EnLiqMDVal, EnLiqMDErr, "Liquid MD", "U/N", "U*")

# Read the data for the gas phase
x, EnGasMCVal, EnGasMCErr = read_data('output_gas_epot_MC.dat')
_, EnGasMDVal, EnGasMDErr = read_data('output_gas_epot_MD.dat')

# Calculate the number of throws
M = L * N
x *= L

# Plot the gas phase data
plot_data(x, EnGasMCVal, EnGasMCErr, "Gas MC", EnGasMDVal, EnGasMDErr, "Gas MD", "U/N", "U*")


```

```python
x, EnSolMCVal, EnSolMCErr = read_data('output_solid_pres_MC.dat')
_, EnSolMDVal, EnSolMDErr = read_data('output_solid_pres_MD.dat')

x *= L

# Plot the gas phase data
plot_data(x, EnSolMCVal, EnSolMCErr, "Solid MC", EnSolMDVal, EnSolMDErr, "Solid MD", "P", "P*")


x, EnSolMCVal, EnSolMCErr = read_data('output_liquid_pres_MC.dat')
_, EnSolMDVal, EnSolMDErr = read_data('output_liquid_pres_MD.dat')

x *= L

# Plot the data
plot_data(x, EnSolMCVal, EnSolMCErr, "Liquid MC", EnSolMDVal, EnSolMDErr, "Liquid MD", "P", "P*")


x, EnSolMCVal, EnSolMCErr = read_data('output_gas_pres_MC.dat')
_, EnSolMDVal, EnSolMDErr = read_data('output_gas_pres_MD.dat')

x *= L

# Plot the data
plot_data(x, EnSolMCVal, EnSolMCErr, "Gas MC", EnSolMDVal, EnSolMDErr, "Gas MD", "P", "P*")
```

```python
def read_gr(filename):
    # Read the data from file
    data = np.loadtxt(filename)
    # Extract the values and errors as separate Series
    x = data[:, 0]
    val = data[:, 1]
    err = data[:, 2]
    return x, val, err

def plot_gr(x, y1_val, y1_err, y1_label, y2_val, y2_err, y2_label):
    # Set up the plot
    plt.errorbar(x, y1_val, yerr=y1_err, label=y1_label)
    plt.errorbar(x, y2_val, yerr=y2_err, label=y2_label, alpha=0.7)
    plt.xlabel('r')
    plt.ylabel('g(r)')
    plt.title('Radial Distribution Function')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_gr_diff(x, y1_val, y1_err, y2_val, y2_err):
    # Set up the plot
    plt.errorbar(x, y1_val - y2_val, yerr=(y1_err**2+y2_err**2)**0.5)
    plt.xlabel('r')
    plt.ylabel('g(r)')
    plt.title('G(r) difference between MC and MD')
    plt.grid(True)
    plt.show()    
    
r, MC, MCerr = read_gr('output_solid_gr_MC.dat') 
r, MD, MDerr = read_gr('output_solid_gr_MD.dat') 
plot_gr(r, MC, MCerr, 'Solid MC', MD, MDerr, 'Solid MD')
```

```python
r, MC, MCerr = read_gr('output_liquid_gr_MC.dat') 
r, MD, MDerr = read_gr('output_liquid_gr_MD.dat') 
plot_gr(r, MC, MCerr, 'Liquid MC', MD, MDerr, 'Liquid MD')
```

```python
r, MC, MCerr = read_gr('output_gas_gr_MC.dat') 
r, MD, MDerr = read_gr('output_gas_gr_MD.dat') 
plot_gr(r, MC, MCerr, 'Gas MC', MD, MDerr, 'Gas MD')
```

### Reusults
We see that for the liquid and the gas the results obtained with the Monte Carlo simulations aand the Molecular Dynamics simulations are perfectly compatible, while in the solid there is a really small difference, although it is big in relations to the standard deviations. There are two resons why this should't worry us too much: first of all the two simulations should converge when the number of particles goes to infinity, and we only have 108 of them; second the MC simulation reproduces a system with a set temperature, while the energy of the system might fluctuate a bit, while in the MD the energy is preserved and it's the temperature that fluctuates slightly.
