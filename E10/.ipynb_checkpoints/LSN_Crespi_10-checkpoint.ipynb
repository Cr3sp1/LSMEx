### Note
We chose the first option. We are using a population of 50 in every node, for 6 nodes.

```python
import numpy as np
import matplotlib.pyplot as plt

AvgCon = np.loadtxt('con.bestHalf.dat')
AvgUnc = np.loadtxt('unc.bestHalf.dat')
BestCon = np.loadtxt('con.best.dat')
BestUnc = np.loadtxt('unc.best.dat')

AvgC = AvgCon[:, 1]
AvgU = AvgUnc[:, 1]
BestC = BestCon[:, 1]
BestU = BestUnc[:, 1]

# Generate x-axis values
x = AvgCon[:, 0]

plt.plot(x, AvgC, label='Connected')
plt.plot(x, AvgU, label='Independent')
plt.xlabel('Generation')
plt.ylabel('<L1> of best half of population')
plt.title('<L1> Comparison')
plt.legend()
plt.grid(True)
plt.show()

plt.plot(x, BestC, label='Connected')
plt.plot(x, BestU, label='Independent')
plt.xlabel('Generation')
plt.ylabel('L1 Comparison')
plt.title('Evolution of L1 of the best path')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Read the data file
data_file = 'unc.path.dat'
with open(data_file, 'r') as f:
    lines = f.readlines()

# Extract x and y coordinates
x = []
y = []
for line in lines:
    values = line.split()
    x.append(float(values[0]))
    y.append(float(values[1]))

# Add the first point at the end to connect the last point with the first one
x.append(x[0])
y.append(y[0])

plt.figure(figsize=(8,5))
# Plot the points and connect them
plt.plot(x, y, marker='o')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Best Path for Independent Search')
plt.grid(True)
plt.show()

# Read the data file
data_file = 'con.path.dat'
with open(data_file, 'r') as f:
    lines = f.readlines()

# Extract x and y coordinates
x = []
y = []
for line in lines:
    values = line.split()
    x.append(float(values[0]))
    y.append(float(values[1]))

# Add the first point at the end to connect the last point with the first one
x.append(x[0])
y.append(y[0])

# Plot the points and connect them
plt.figure(figsize=(8,5))
plt.plot(x, y, marker='o', color = 'C1')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Best Path for Connected Search')
plt.grid(True)
plt.show()
```

### Results
As we can see, with a relatively low population of 50 paths, the connected search performa much better, both in terms of speed of convergence and best path reached. While the indipendent search is equivalent to running the algorithm 6 times and taking the best result, the connected search is more akin to having an incresed population of 300 (of course, it's not the same since the amount of genetic material shared is limited). Multiplying the poulation by a certain amount is much more beneficial than running the algorithm that many times, as well as more computationally costly, since the complexity of generating a new population is at least O(nlogn), due to the fact that it requires sorting the population. Taking this into account explains why the connected search gives better results while not requiring a meaningful increase in computational power.
